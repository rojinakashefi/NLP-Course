## Preprocessing on data

```python
def preprocessing(file_name):
  sentences = []
  with open(file_name) as file:
    for line in file:
      line = re.sub(r"[\s]{2, }", " ", line)
      line = line.strip()
      words = line.split(" ")
      words = [x for x in words if x!='']
      if len(words) > 2 :
        sentences.append(words)
  return sentences
```

```python
sentences = preprocessing('train.txt')
len(sentences)
```

### Unigram
In unigram we count the occurance of the word in whole text.

```python
word_count = 0
each_word_count = {}
for sentence in sentences:
  for word in sentence:
    word_count +=1
    if word in each_word_count.keys():
      each_word_count[word] += 1
    else:
      each_word_count[word] = 1
```

### Bigram
In bigram we count the occurance of two word next to each other in whole text.

**P( wn | wn-1) = C( wn, wn-1 ) / C( wn-1 )**

```python
bigram_word_count = {}
for sentence in sentences:
  for i in range(0, len(sentence)):
    if sentence[i] not in bigram_word_count.keys():
      bigram_word_count[sentence[i]] = []
    if i-1>=0 :
      bigram_word_count[sentence[i]].append(sentence[i-1])
```

#### Using nltk counter

In previous method we only used each word list to calculate bigrams, but here because we are using NLTK library we can process the whole text and its much accurate.

```python
from nltk import FreqDist
from nltk.util import ngrams
import collections
```

```python
ngramfreqdist = FreqDist()
for sentence in sentences:
  sen_ngram = ngrams(sentence,2)
  ngramfreqdist.update(sen_ngram)
count_ngrams = collections.Counter(ngramfreqdist)
```

```python
for rank,selected_ngram in enumerate(count_ngrams.most_common(20)):
  items,counter = selected_ngram
  print(f'{rank + 1} : {items} {counter}')
```

```python
def compute_freq(file:str,n:int):
  sentences = preprocessing('train.txt')
  all_ngrams = []
  for sentence in sentences:
    ngrams = zip(*[sentence[i:] for i in range(n)])
    all_ngrams.extend([" ".join(ngram) for ngram in ngrams])
  return all_ngrams,len(all_ngrams)
```

### NLTK

```python
from nltk.util import ngrams
from nltk import FreqDist
```

```python
def compute_freq(file_name:str,n:int):
  all_ngrams = []
  sentences_ = preprocessing(file_name)
  for sentence in sentences_:
    all_ngrams.extend(list(ngrams(sentence,n)))
  return all_ngrams, len(all_ngrams)
```
